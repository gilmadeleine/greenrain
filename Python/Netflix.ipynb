```python
import csv
import pandas as pd
f=open("netflix.csv")
data = csv.reader(f)
header=next(data)
list=[]
for row in data:
    print(row[-3]) 
    list.append(row[-3])
print(list)
```

```python
df = pd.DataFrame({"ratings":list})
df
```

```python
search_keyword = ['PG', 'MA','R','14','Y7','G']

for keyword in search_keyword:
    df[keyword] = df["ratings"].str.contains(keyword)
```

```python
df.to_csv("netflixratings.csv")
```
